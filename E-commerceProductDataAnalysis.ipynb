```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/BigBasket Products.csv 22-25-03-354.csv')
```

```python
# Check missing values
df.isnull().sum()
```

```python
df['rating'] = df['rating'].fillna(df['rating'].mean())
df['brand'] = df['brand'].fillna('Unknown')
df['product'] = df['product'].fillna('Unknown')
```

```python
data = df
title_style = {
    'fontsize': 12,
    'fontweight': 'bold',
    'color': 'darkblue',
    'loc': 'center'
}
```

```python
plt.figure(figsize=(8,5))
sns.histplot(data=data , x = 'rating' , bins = 50 , kde = True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,8))

sns.scatterplot(data=data , x = 'market_price', y = 'sale_price' , palette='Set3')
plt.title('Relation of Market_Price and Sale_price' ,**title_style)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,9))

df['category'].value_counts().head(10).plot(kind='bar')
plt.title("Top category",**title_style)
plt.xticks(rotation =45 )
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,6))

sns.histplot(data=data,x='rating' , bins = 20 , kde = True)
plt.title("Ratings Distribution" , **title_style)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,6))

df['brand'].value_counts().head(10).plot(kind='bar')
plt.title("Top Brand" , **title_style)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,7))

sns.boxplot(data=data , x = 'category' , y = 'sale_price')
plt.xticks(rotation=45)
plt.title("Price Distribution by Category",**title_style)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.scatter(data=data , x ='sale_price' , y = 'rating' , alpha=0.2)
plt.title("Price vs Rating",**title_style)
plt.xlabel("sale_price")
plt.ylabel("rating")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.heatmap(df[['sale_price','market_price','rating']].corr(), annot=True)
plt.title("Correlation Matrix")
plt.show()
```
